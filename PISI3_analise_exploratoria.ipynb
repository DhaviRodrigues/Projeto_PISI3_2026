```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/Imdb Movies Dataset.csv', on_bad_lines='skip')

df_aux = df.copy()
media_votos = df_aux['vote_count'].mean()

df_nicho = df_aux[
    (df_aux['vote_count'] < media_votos) &
    (df_aux['vote_count'] > 20) &
    (df_aux['runtime'] > 0)
].copy()

df.head(10)
```

```python
#Exibe
df_aux['release_year'] = pd.to_datetime(df_aux['release_date'], errors='coerce').dt.year
df_plot = df_aux.dropna(subset=['release_year']).copy()
df_plot['release_year'] = df_plot['release_year'].astype(int)

df_ano = df_plot.groupby('release_year')['popularity'].mean().reset_index()

df_ano['popularity_smooth'] = df_ano['popularity'].rolling(window=5, min_periods=1).mean()

plt.figure(figsize=(15, 7))

plt.fill_between(df_ano['release_year'], df_ano['popularity_smooth'], color="skyblue", alpha=0.4)
plt.plot(df_ano['release_year'], df_ano['popularity_smooth'], color="violet", alpha=0.6, linewidth=2)

plt.title('Tendência Histórica de Popularidade dos Filmes', fontsize=16)
plt.xlabel('Ano de Lançamento', fontsize=12)
plt.ylabel('Popularidade Média', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)

plt.xticks(df_ano['release_year'][::5], rotation=45)

plt.tight_layout()
plt.show()
```

```python
#Exibe os gêneros com maior número de instâncias
generos_limpos = df['genres'].str.split(',').explode().str.strip()

generos_limpos.value_counts().head(19).plot(kind='bar', figsize=(11,7), color='skyblue')

plt.title('Ranking do Número de Filmes por Gênero')
plt.ylabel('Quantidade de Instâncias')
plt.xlabel('Gênero')
plt.xticks(fontsize=10)
plt.show()
```

```python
#Exibe um gráfico que representa a relação entre notas totais com média de notas para ano de lançamento

minimo_votos = df_plot['vote_count'].mean()

df_filtrado = df_plot[df_plot['vote_count'] >= (minimo_votos - 12)]

plt.figure(figsize=(12, 6))
df_filtrado.groupby('release_year')['vote_average'].mean().plot(kind='line', marker='o',linewidth=2, label='Média de Notas por Ano')

plt.title(f'Média de Notas por Ano')
plt.xlabel('Ano de Lançamento')
plt.ylabel('Média de Notas')
plt.grid(True, alpha=0.3)

plt.axhline(y=df_filtrado['vote_average'].mean(), color='red', linestyle='--', alpha=0.5, label='Média Global de Notas Anuais')
plt.legend(loc='best')

plt.show()
```

```python
#Exibe um wordcloud que mostra a maior frequência de palavras-chave e lista as suas top 50
from wordcloud import WordCloud

keywords_series = df_aux['keywords'].dropna().str.split(',').explode().str.strip()


frequencia_keywords = keywords_series.value_counts().to_dict()

wordcloud = WordCloud(
    width=1000,
    height=500,
    background_color='white',
    colormap='viridis',
    max_words=100
).generate_from_frequencies(frequencia_keywords)

plt.figure(figsize=(15, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title('Nuvem de Palavras-Chave dos Filmes', fontsize=20)
plt.show()

print(keywords_series.value_counts().head(50))
```

```python
#Exibe um gráfico que mostra a relação entre duração do filme e ano de lançamento

df_aux['release_year'] = pd.to_datetime(df_aux['release_date'], errors='coerce').dt.year
df_aux['release_year'] = pd.to_datetime(df_aux['release_date'], errors='coerce').dt.year

df_runtime = df_aux.dropna(subset=['release_year', 'runtime']).copy()

df_runtime = df_runtime[(df_runtime['release_year'] >= 1900) &
                        (df_runtime['release_year'] <= 2023) &
                        (df_runtime['runtime'] > 10) &
                        (df_runtime['runtime'] < 300)]

df_runtime['release_year'] = df_runtime['release_year'].astype(int)

runtime_medio_ano = df_runtime.groupby('release_year')['runtime'].mean()

plt.figure(figsize=(18, 10))

plt.scatter(df_runtime['release_year'], df_runtime['runtime'],
            alpha=0.1, s=10, color='gray', label='Filmes individuais')

plt.plot(runtime_medio_ano.index, runtime_medio_ano.values,
         color='blue', linewidth=3, label='Duração Média Anual')

plt.title('Evolução da Duração dos Filmes (Runtime) ao Longo dos Anos', fontsize=15)
plt.xlabel('Ano de Lançamento')
plt.ylabel('Duração (minutos)')
plt.grid(True, alpha=0.3)
plt.ylim(0, 250)

lgnd = plt.legend(loc='upper left', scatterpoints=1, fontsize=10)
for handle in lgnd.legend_handles:
    handle.set_alpha(1.0)
    if hasattr(handle, 'set_sizes'):
        handle.set_sizes([50])

plt.show()
```

```python
#Cria um gráfico de boxplot que verifica a relação de filmes nichados, média de avaliação e tempo de filme

bins = [0, 40, 90, df_nicho['runtime'].max()]
labels = ['Curtas (<40m)', 'Médios (40-90m)', 'Longas (>90m)']

df_nicho['duration_category'] = pd.cut(df_nicho['runtime'], bins=bins, labels=labels)

plt.figure(figsize=(12, 7))
sns.set_style("whitegrid")

plot = sns.boxplot(
     data=df_nicho,
    x='duration_category',
    y='vote_average',
    hue='duration_category',
    palette='pastel',
    legend=False
)

plt.title('Distribuição de Notas por Duração (Filmes de Nicho)', fontsize=16)
plt.xlabel('Categorias de Tempo de Tela', fontsize=12)
plt.ylabel('Nota Média', fontsize=12)

plt.axhline(df_nicho['vote_average'].mean(), color='r', linestyle='--', alpha=0.6, label='Média Geral')

plt.tight_layout()
plt.show()

print("Quantidade de filmes por categoria:")
print(df_nicho['duration_category'].value_counts().sort_index())
```

```python
#Cria um gráfico de violino que verifica a relação de filmes nichados, média de avaliação e tempo de filme

plt.figure(figsize=(12, 8))
sns.set_style("whitegrid")

bins = [0, 40, 90, df_nicho['runtime'].max()]
labels = ['Curtas (<40m)', 'Médios (40-90m)', 'Longas (>90m)']

df_nicho['duration_category'] = pd.cut(df_nicho['runtime'], bins=bins, labels=labels)

ax = sns.violinplot(
    data=df_nicho,
    x='duration_category',
    y='vote_average',
    hue='duration_category',
    palette='muted',
    inner="quartile",
    bw_adjust=0.5,
    legend=False,
)

# APLICAÇÃO DO BOLD NOS TEXTOS DOS VIOLINOS (EIXO X)
ax.set_xticks(range(len(labels)))

# 2. Agora aplicamos os labels com as propriedades de fonte
ax.set_xticklabels(labels, fontweight='bold', fontsize=11)

#plt.title('Densidade de Notas por Duração', fontsize=16)
plt.ylim(0, 11)
plt.xlabel('Categorias de Tempo de Tela', fontsize=12, fontweight='bold')
plt.ylabel('Nota Média', fontsize=12, fontweight='bold')
plt.axhline(df_nicho['vote_average'].mean(), color='r', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

```python
#Realiza a categorização dos países de origem em agrupamentos
df_country_analysis = df_aux.copy()

def parse_country_list(s):
    try:
        if pd.isna(s) or s == '[]' or s == '':
            return []
        return [country.strip() for country in s.split(',')]
    except (ValueError, SyntaxError):
        return []

df_country_analysis['production_countries_list'] = df_country_analysis['production_countries'].apply(parse_country_list)

df_countries_exploded = df_country_analysis.explode('production_countries_list')

df_countries_exploded.dropna(subset=['production_countries_list', 'vote_average', 'vote_count', 'popularity'], inplace=True)

df_countries_exploded = df_countries_exploded[df_countries_exploded['production_countries_list'] != '']

film_counts_per_country = df_countries_exploded.groupby('production_countries_list')['id'].nunique().reset_index()
film_counts_per_country.rename(columns={'id': 'film_count'}, inplace=True)

countries_with_min_films = film_counts_per_country[film_counts_per_country['film_count'] >= 20]['production_countries_list']
df_countries_exploded_filtered_by_film_count = df_countries_exploded[df_countries_exploded['production_countries_list'].isin(countries_with_min_films)]

country_metrics = df_countries_exploded_filtered_by_film_count.groupby('production_countries_list')[['vote_average', 'vote_count', 'popularity']].mean().reset_index()

country_metrics = pd.merge(country_metrics, film_counts_per_country, on='production_countries_list', how='left')
```

```python
#Cria um gráfico de linhas que utiliza gêneros drama, criminal, faroeste e comédia para verificar sua têndencia de 1920 a 2000

df_usa = df_aux[df_aux['production_countries'].str.contains('United States', na=False)].copy()

df_usa['release_date'] = pd.to_datetime(df_usa['release_date'], errors='coerce')
df_usa['year'] = df_usa['release_date'].dt.year

df_usa_evolucao = df_usa[(df_usa['year'] >= 1920) & (df_usa['year'] <= 2000)].copy()

df_usa_evolucao['genre_list'] = df_usa_evolucao['genres'].str.split(', ')

df_exploded = df_usa_evolucao.explode('genre_list')
df_final = df_exploded[df_exploded['genre_list'].isin(['Drama', 'Crime', 'Western', 'Comedy'])].copy()

df_counts = df_final.groupby(['year', 'genre_list']).size().reset_index(name='film_count')

plt.figure(figsize=(14, 7))
sns.set_style("whitegrid")

palette = {'Crime': '#ff7f0e', 'Western': '#2ca02c', 'Drama': '#1f77b4', 'Comedy' : '#B22300'}

dashes = {'Comedy': '', 'Drama': '', 'Crime': (3, 3), 'Western': (3, 3)}

ordem_legenda = ['Drama', 'Comedy', 'Crime', 'Western']

ax = sns.lineplot(
    data=df_counts,
    x='year',
    y='film_count',
    hue='genre_list',
    style='genre_list',
    palette=palette,
    dashes=dashes,
    hue_order=ordem_legenda,
    style_order=ordem_legenda,
    linewidth=2.5
)

plt.title('EUA: Evolução da Produção por Gênero (1920 - 2000)', fontsize=16, fontweight='bold')
plt.xlabel('Ano de Lançamento', fontsize=12, fontweight='bold')
plt.ylabel('Quantidade de Filmes Produzidos', fontsize=12, fontweight='bold')

handles, labels = ax.get_legend_handles_labels()
plt.legend(handles=handles, labels=labels, title='Gêneros', title_fontsize='12', fontsize='11', loc='upper left')

plt.xticks(range(1920, 2005, 5))

plt.tight_layout()
plt.show()
```
