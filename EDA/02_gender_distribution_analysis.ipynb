# Cálculo da Quantidade Total de Filmes e Média Global de Votos de Filme
## Objetivos:
- Usar como parâmetro de Separação entre Filmes mainstream e Filmes alternativos a média global de votos em filmes

- Calcular a quantidade total de filmes para alcançar mais precisão no cálculo da média

```python
df = pd.read_csv('../Imdb_Movie_Dataset.csv')

df.columns = [c.strip().lower() for c in df.columns]
col_votos = 'vote_count' if 'vote_count' in df.columns else [c for c in df.columns if 'vote' in c][0]

df = df.dropna(subset=[col_votos])
df_sorted = df.sort_values(by=col_votos, ascending=False).reset_index(drop=True)
df_sorted['Ranking'] = range(1, len(df_sorted) + 1)
```

```python
import numpy as np

df_filtered = df.dropna(subset=['genres']).copy()
df_filtered['genres'] = df_filtered['genres'].apply(lambda x: x.split(', ') if isinstance(x, str) else [])


media_global_vote_count_all_movies = df['vote_count'].mean()

filmes_mainstream = df_filtered[df_filtered['vote_count'] > media_global_vote_count_all_movies]

filmes_alternativos = df_filtered[(df_filtered['vote_count'] > 0) & (df_filtered['vote_count'] <= media_global_vote_count_all_movies)]


mainstream_movies = filmes_mainstream
long_tail_movies = filmes_alternativos

total_movies_in_df = len(df)
print(f"\nTotal de filmes: {total_movies_in_df:,}")
print(f"Média Global de votos: {media_global_vote_count_all_movies:.2f} votos")
```

# Limpeza do dataset antes da geração dos gráficos

```python
mainstream_exploded = mainstream_movies.explode('genres')
df_mainstream_genres = mainstream_exploded['genres'].value_counts().reset_index()
df_mainstream_genres.columns = ['Genre', 'Count']

long_tail_exploded = long_tail_movies.explode('genres')
df_long_tail_genres = long_tail_exploded['genres'].value_counts().reset_index()
df_long_tail_genres.columns = ['Genre', 'Count']

print("Gêneros Mainstream (Top 10):")
print(df_mainstream_genres.head(10))
print("\nGêneros Alternativos (Top 10):")
print(df_long_tail_genres.head(10))
```

# Gera os Gráficos de comparação entre Filmes Mainstream e Filmes Alternativos

```python
import matplotlib.pyplot as plt
import seaborn as sns

N = 10 

global_mean_vote_count_display = round(media_global_vote_count_all_movies)

fig, axes = plt.subplots(2, 1, figsize=(12, 12))
fig.suptitle('Top 10 Gêneros de filmes entre Filmes Mainstream vs. Filmes Alternativos', fontsize=16, y=1.02)

max_count = max(df_mainstream_genres['Count'].max(), df_long_tail_genres['Count'].max())

sns.barplot(x='Count', y='Genre', data=df_mainstream_genres.head(N), ax=axes[0], palette='viridis')
axes[0].set_title(f'Top {N} Gêneros de filmes Mainstream (Média Global de Votos > {global_mean_vote_count_display})')
axes[0].set_xlabel('Quantidade de filmes')
axes[0].set_ylabel('Gênero')
axes[0].tick_params(axis='x', rotation=0)
axes[0].set_xlim(0, max_count) 

sns.barplot(x='Count', y='Genre', data=df_long_tail_genres.head(N), ax=axes[1], palette='viridis') 
axes[1].set_title(f'Top {N} Gêneros de Filmes Alternativos (0 < Média Global de Votos <= {global_mean_vote_count_display})')
axes[1].set_xlabel('Quantidade de Filmes')
axes[1].set_ylabel('Gênero')
axes[1].tick_params(axis='x', rotation=0)
axes[1].set_xlim(0, max_count) 

plt.tight_layout(rect=[0, 0.03, 1, 0.98])
plt.show()
```
