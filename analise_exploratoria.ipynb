```python
import pandas as pd


df = pd.read_csv('Imdb Movies Dataset.csv', on_bad_lines='skip')
df_aux = df.copy()

df.head(10)
```

```python
#Dados sobre filmes "estrangeiros"

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df_country_analysis = df_aux.copy()

def parse_country_list(s):
    try:
        if pd.isna(s) or s == '[]' or s == '':
            return []
        return [country.strip() for country in s.split(',')]
    except (ValueError, SyntaxError):
        return []

df_country_analysis['production_countries_list'] = df_country_analysis['production_countries'].apply(parse_country_list)

df_countries_exploded = df_country_analysis.explode('production_countries_list')

df_countries_exploded.dropna(subset=['production_countries_list', 'vote_average', 'vote_count', 'popularity'], inplace=True)

df_countries_exploded = df_countries_exploded[df_countries_exploded['production_countries_list'] != '']

film_counts_per_country = df_countries_exploded.groupby('production_countries_list')['id'].nunique().reset_index()
film_counts_per_country.rename(columns={'id': 'film_count'}, inplace=True)

countries_with_min_films = film_counts_per_country[film_counts_per_country['film_count'] >= 20]['production_countries_list']
df_countries_exploded_filtered_by_film_count = df_countries_exploded[df_countries_exploded['production_countries_list'].isin(countries_with_min_films)]

country_metrics = df_countries_exploded_filtered_by_film_count.groupby('production_countries_list')[['vote_average', 'vote_count', 'popularity']].mean().reset_index()

country_metrics = pd.merge(country_metrics, film_counts_per_country, on='production_countries_list', how='left')

if not country_metrics.empty:
    print('Top 20 Países por Média de Votos:')
    display(country_metrics.sort_values(by='vote_average', ascending=False).head(20))

    print('\nTop 20 Países por Média de Popularidade:')
    display(country_metrics.sort_values(by='popularity', ascending=False).head(20))

    print('\nTop 20 Países por Média de Contagem de Votos:')
    display(country_metrics.sort_values(by='vote_count', ascending=False).head(20))

    print('\nTop 20 Países por Contagem de Filmes:')
    display(country_metrics.sort_values(by='film_count', ascending=False).head(20))

    # Visualização: Top 20 Países por Média de Votos
    plt.figure(figsize=(15, 6))
    plt.subplot(1, 2, 1)
    sns.barplot(x='production_countries_list', y='vote_average', data=country_metrics.nlargest(20, 'vote_average'), palette='viridis')
    plt.title('Top 20 Países por Média de Votos')
    plt.xlabel('País de Produção')
    plt.ylabel('Média de Votos')
    plt.xticks(rotation=45, ha='right')

    # Visualização: Top 20 Países por Média de Contagem de Votos
    plt.subplot(1, 2, 2)
    sns.barplot(x='production_countries_list', y='vote_count', data=country_metrics.nlargest(20, 'vote_count'), palette='plasma')
    plt.title('Top 20 Países por Média de Contagem de Votos')
    plt.xlabel('País de Produção')
    plt.ylabel('Média de Contagem de Votos')
    plt.xticks(rotation=45, ha='right')

    plt.tight_layout()
    plt.show()
else:
    print("Não há dados suficientes para gerar os gráficos após a filtragem. O DataFrame 'country_metrics' está vazio.")
```

```python
#Filmes de faroeste ao longo do tempo e análise de gêneros mais e menos recorrentes

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


df_genres_analysis = df_aux.copy()
df_genres_analysis['release_year'] = pd.to_datetime(df_genres_analysis['release_date'], errors='coerce').dt.year
df_genres_analysis.dropna(subset=['genres', 'release_year'], inplace=True)

df_genres_exploded = df_genres_analysis.assign(genres=df_genres_analysis['genres'].str.split(',')).explode('genres')
df_genres_exploded['genres'] = df_genres_exploded['genres'].str.strip()

western_films = df_genres_exploded[df_genres_exploded['genres'] == 'Western']

western_pre_2000 = western_films[western_films['release_year'] < 2000]
western_post_2000 = western_films[western_films['release_year'] >= 2000]

print(f"Filmes de faroeste antes do ano 2000: {len(western_pre_2000)} filmes")
print(f"Filmes de faroeste a partir do ano 2000: {len(western_post_2000)} filmes")

western_yearly_counts = western_films.groupby('release_year').size().reset_index(name='count')

plt.figure(figsize=(15, 6))
sns.lineplot(x='release_year', y='count', data=western_yearly_counts, errorbar=None)
plt.title('Número de Filmes de faroeste por Ano de Lançamento')
plt.xlabel('Ano de Lançamento')
plt.ylabel('Número de Filmes')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# --- Análise de gêneros mais e menos recorrentes ao longo do tempo ---
# Contagem de filmes por gênero e por ano
genre_yearly_counts = df_genres_exploded.groupby(['release_year', 'genres']).size().reset_index(name='count')

min_films_per_year = 100
year_total_counts = genre_yearly_counts.groupby('release_year')['count'].sum()
valid_years = year_total_counts[year_total_counts >= min_films_per_year].index
genre_yearly_counts_filtered = genre_yearly_counts[genre_yearly_counts['release_year'].isin(valid_years)]

genre_yearly_proportions = genre_yearly_counts_filtered.copy()
genre_yearly_proportions['proportion'] = genre_yearly_proportions.groupby('release_year')['count'].transform(lambda x: x / x.sum())

top_genres = df_genres_exploded['genres'].value_counts().head(10).index

print("\n--- Tendência dos Top 10 Gêneros Mais Comuns ---")
plt.figure(figsize=(18, 10))
for genre in top_genres:
    genre_data = genre_yearly_proportions[genre_yearly_proportions['genres'] == genre]
    if not genre_data.empty:
        sns.lineplot(x='release_year', y='proportion', data=genre_data, label=genre, errorbar=None)

plt.title('Proporção de Filmes por Gênero ao Longo do Tempo (end-2023')
plt.xlabel('Ano de Lançamento')
plt.ylabel('Proporção de Filmes')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

period_start = genre_yearly_proportions['release_year'].min()
period_end = 2023.0 

start_proportions = genre_yearly_proportions[genre_yearly_proportions['release_year'].between(period_start, period_start + 4)].groupby('genres')['proportion'].mean().fillna(0)
end_proportions = genre_yearly_proportions[genre_yearly_proportions['release_year'].between(period_end - 4, period_end)].groupby('genres')['proportion'].mean().fillna(0)

proportion_change = (end_proportions - start_proportions).sort_values(ascending=False)

print("\n--- Gêneros com Maior Aumento na Proporção (Comparando os 5 anos mais recentes com os 5 anos mais antigos) ---")
display(proportion_change.head(10))

print("\n--- Gêneros com Maior Diminuição na Proporção (Comparando os 5 anos mais recentes com os 5 anos mais antigos) ---")
display(proportion_change.tail(10))
```

```python
#Duração do filme relacionado a quantidade de votos

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df_runtime_analysis = df_aux.copy()

df_runtime_analysis.replace({'runtime': 0}, np.nan, inplace=True)
df_runtime_analysis.dropna(subset=['runtime', 'vote_average', 'vote_count'], inplace=True)

df_runtime_analysis = df_runtime_analysis[df_runtime_analysis['vote_count'] > 10]

print("Análise da Relação entre Duração do Filme (Runtime) e Média de Votos:")

plt.figure(figsize=(12, 7))
sns.scatterplot(x='runtime', y='vote_average', data=df_runtime_analysis, alpha=0.3, s=10)
plt.title('Duração do Filme vs. Média de Votos')
plt.xlabel('Duração (minutos)')
plt.ylabel('Média de Votos')
plt.grid(True, alpha=0.5)
plt.show()

bins_runtime = [0, 90, 120, 150, 180, 240, df_runtime_analysis['runtime'].max()]
labels_runtime = ['<90 min', '90-120 min', '120-150 min', '150-180 min', '180-240 min', '>240 min']

df_runtime_analysis['runtime_category'] = pd.cut(
    df_runtime_analysis['runtime'],
    bins=bins_runtime,
    labels=labels_runtime,
    right=True,
    include_lowest=True
)

average_vote_by_runtime = df_runtime_analysis.groupby('runtime_category')['vote_average'].mean().reset_index()
print("\nMédia de Votos por Categoria de Duração:")
display(average_vote_by_runtime)

plt.figure(figsize=(10, 6))
sns.barplot(x='runtime_category', y='vote_average', data=average_vote_by_runtime, palette='viridis')
plt.title('Média de Votos por Categoria de Duração')
plt.xlabel('Duração do Filme')
plt.ylabel('Média de Votos')
plt.ylim(df_runtime_analysis['vote_average'].min() - 0.5, df_runtime_analysis['vote_average'].max() + 0.5)
plt.grid(axis='y', alpha=0.75)
plt.show()
```
